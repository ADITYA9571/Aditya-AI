File Input

```python
file_name = input("Enter file name: ")

with open(file_name, "r", encoding="utf-8") as file:
    content = file.read()

print(content)
```

Exception Handling

```python
file_name = input("Enter file name: ")

try:
    with open(file_name, "r", encoding="utf-8") as file:
        content = file.read()
        print(content)

except FileNotFoundError:
    print("File not found")
```

```python
def count_words(text):
    return len(text.split())

print(count_words("Hello world"))
```
